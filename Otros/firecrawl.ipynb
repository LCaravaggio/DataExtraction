```python
from firecrawl import FirecrawlApp
from google.colab import userdata

API_key=userdata.get('FIRECRAWL')

app = FirecrawlApp(api_key=API_key)

result = app.scrape(
    "https://www.zonaprop.com.ar/departamentos-venta-flores.html",
    formats=["markdown"]
)

print(result)
```
