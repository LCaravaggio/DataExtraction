# Bajar base entera

```python
from google.colab import userdata
API_key=userdata.get('ALPHACAST')
```

```python
import requests
import pandas as pd
from io import StringIO

url_base='https://api.alphacast.io/datasets'
dataset='5359'
url = url_base + f'/{dataset}/data?apiKey={API_key}&$format=csv'

r1 = requests.get(url)
```

```python
data = StringIO(r1.text)
df = pd.read_csv(data, delimiter=',')
```

```python
df
```
