```python
import requests
import urllib.parse
API_BASE_URL = "http://apis.datos.gob.ar/series/api/"

def get_api_call(ids, **kwargs):
    kwargs["ids"] = ",".join(ids)
    return "{}{}?{}".format(API_BASE_URL, "series", urllib.parse.urlencode(kwargs))

api_call = get_api_call(["143.3_NO_PR_2004_A_21"], start_date="2020-01")
print(api_call)
```

```python
datos=requests.get(api_call).json()
```

```python
import pandas as pd
pd.DataFrame(datos.get("data"), columns=['Fecha', 'EMAE'])
```
